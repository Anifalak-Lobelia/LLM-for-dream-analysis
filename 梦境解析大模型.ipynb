### 3.3.2 文本处理
下面的代码中定义了`ChineseTextSplitter`的类，它继承自`CharacterTextSplitter`。我们的类是一个高度专门化的工具，用于处理和优化中文文本数据，特别是从不同格式的文档中提取和准备文本数据以供进一步的分析和处理。具体来说做了以下处理：

- **去除英文单词**：使用正则表达式从文本中删除所有英文字母。这一步骤是为了减少处理不相关语言的复杂性，特别是在主要处理中文文本时。

- **检查无意义的文本**：通过正则表达式检测文本是否只包含数字、符号或空白字符，并确认是否缺少中文字符。这有助于排除不含实际信息内容的文本，从而提高数据的质量和后续处理的效率。

- **基于字符数分割**：根据预设的最大句子长度（`sentence_size`）对文本进行分割，确保处理的文本块大小适中，既不过长以致难以处理，也不过短而丢失上下文信息。

- **细致的分句**：对较长的文本行进行更细致的分割，使用一系列复杂的正则表达式来正确地切分句子、短语和段落。这包括处理标点符号、引号和省略号等，确保文本的自然语义边界得到尊重。

```python
class ChineseTextSplitter(CharacterTextSplitter):
    def __init__(self, pdf: bool = False, sentence_size: int = SENTENCE_SIZE, **kwargs):
        super().__init__(**kwargs)
        self.pdf = pdf
        self.sentence_size = sentence_size

    def preprocess_text(self, text: str) -> str:
        # 去除所有英文单词
        text = re.sub(r'[a-zA-Z]', '', text)
        return text

    def is_meaningless(self, text: str) -> bool:
        # 判断文本是否只包含符号、空格、数字等，或没有中文字符
        contains_chinese = re.search(r'[\u4e00-\u9fff]', text) is not None
        meaningless = re.fullmatch(r'[\W_0-9\s]*', text) is not None
        return meaningless or not contains_chinese

    def split_text(self, text: str) -> List[str]:
        if self.pdf:
            text = re.sub(r"\n{3,}", r"\n", text)
            text = re.sub('\s', " ", text)
            text = re.sub("\n\n", "", text)
        
        ret = []
        texts = text.split("\n")
        for text in texts:
            if text.strip() == "":
                continue
            
            # 预处理文本
            text = self.preprocess_text(text)
            # 删除长度小于7个字符的文本
            if len(text) < 7:
                continue

            # 删除只包含符号、空格、数字或没有中文字符的文本
            if self.is_meaningless(text):
                continue
            
            if len(text) > self.sentence_size:
                ret_texts = self.split_mini_text(text)
                merge_texts = []
                start_idx = 0
                for i, text in enumerate(ret_texts):
                    if len("".join(ret_texts[start_idx:i+1])) > self.sentence_size:
                        merge_texts.append("".join(ret_texts[start_idx: i]))
                        start_idx = i
                merge_texts.append("".join(ret_texts[start_idx: ]))
                ret.extend(merge_texts)
            else:
                ret.append(text)
        return ret

    def split_mini_text(self, text: str) -> List[str]:
        if self.pdf:
            text = re.sub(r"\n{3,}", r"\n", text)
            text = re.sub('\s', " ", text)
            text = re.sub("\n\n", "", text)

        text = re.sub(r' +', '', text)
        text = re.sub(r'([;；.!?。！？\?])([^”’"])', r"\1\n\2", text)
        text = re.sub(r'(\.{6})([^"’”」』])', r"\1\n\2", text)
        text = re.sub(r'(\…{2})([^"’”」』])', r"\1\n\2", text)
        text = re.sub(r'([;；!?。！？\?]["’”」』]{0,1,2})([^;；!?，。！？\?])', r'\1\n\2', text)
        text = text.rstrip()
        ls = [i for i in text.split("\n") if i]
        
        for ele in ls: 
            if len(ele) > self.sentence_size:
                ele1 = re.sub(r'([,，.]["’”」』]{0,2})([^,，.])', r'\1\n\2', ele)
                ele1_ls = ele1.split("\n")
                for ele_ele1 in ele1_ls:
                    if len(ele_ele1) > self.sentence_size:
                        ele_ele2 = re.sub(r'([\n]{1,}| {2,}["’”」』]{0,2})([^\s])', r'\1\n\2', ele_ele1)
                        ele2_ls = ele_ele2.split("\n")
                        for ele_ele2 in ele2_ls:
                            if len(ele_ele2) > self.sentence_size:
                                ele_ele3 = re.sub('( ["’”」』]{0,2})([^ ])', r'\1\n\2', ele_ele2)
                                ele2_id = ele2_ls.index(ele_ele2)
                                ele2_ls = ele2_ls[:ele2_id] + [i for i in ele_ele3.split("\n") if i] + ele2_ls[ele2_id + 1:]
                        ele_id = ele1_ls.index(ele_ele1)
                        ele1_ls = ele1_ls[:ele_id] + [i for i in ele2_ls if i] + ele1_ls[ele_id + 1:]

                id = ls.index(ele)
                ls = ls[:id] + [i for i in ele1_ls if i] + ls[id + 1:]
        return ls

# 加载知识库中的所有.txt文件并划分段落
def load_and_split_text_files(directory):
    texts = []
    splitter = ChineseTextSplitter()
    for filename in os.listdir(directory):
        if filename.endswith('.txt'):
            with open(os.path.join(directory, filename), 'r', encoding='utf-8') as file:
                content = file.read()
                paragraphs = splitter.split_text(content)
                texts.extend(paragraphs)
    return texts
```

### 3.2.3 生成embedding
下面的函数负责将输入的文本转换成embedding。具体来说使用TextEmbedding.call方法调用通义千问预训练embedding模型（text-embedding-v2），将文本输入转换为embedding。

```python
# 生成embedding的函数
def generate_embeddings(input_text):
    rsp = TextEmbedding.call(model='text-embedding-v2',
                             input=input_text,
                             baseurl=baseurl)
    embeddings = [record['embedding'] for record in rsp.output['embeddings']]
    return embeddings if isinstance(input_text, list) else embeddings[0]
```

### 3.2.4 存储embedding至向量数据库
下面的代码实现了将通过调用`generate_embeddings`函数将文本转为embedding并将embedding存储在向量数据库中。这里使用的是FAISS（Facebook AI Similarity Search）库来管理和索引这些向量。

```python
# 将文本区块转为嵌入向量并存储在向量库中
def build_vector_base(texts, vector_base_dir, name):
    os.makedirs(vector_base_dir, exist_ok=True)
    # 生成所有文本段落的嵌embedding
    embeddings = []
    for text in tqdm(texts, desc="文本块向量化中："):
        if not text.strip():
            continue
        embedding = generate_embeddings(text)
        if embedding:
            embeddings.append(embedding)

    # 创建FAISS索引
    if embeddings:
        dimension = len(embeddings[0])
        index = faiss.IndexFlatL2(dimension)
        # 将embedding添加到FAISS索引中
        index.add(np.array(embeddings).astype('float32'))
        # 将索引存储到文件中
        faiss.write_index(index, os.path.join(vector_base_dir, name))
    else:
        print("没有文本生成")
```

将当前目录下传统文学知识库（'knowledge_base/metaphysic'）和心理学知识库（'knowledge_base/psychology'）的内容进行原始文本的划分（`load_and_split_text_files`）、转为embedding（`generate_embeddings`）并存储在向量数据库（`build_vector_base`）的过程。

```python
meta_knowledge_base_dir = 'knowledge_base/metaphysic'
psy_knowledge_base_dir = 'knowledge_base/psychology'
vector_base_dir = 'vector_base'

meta_name = "metaphysic_index"
psy_name = "psychology_index"

# 传统文化解梦知识库向量化
texts = load_and_split_text_files(meta_knowledge_base_dir)
print(f"从传统知识库中载入 {len(texts)} 个文本分块.")
build_vector_base(texts, vector_base_dir, meta_name)
print(f'已完成传统向量库构建，共包含 {len(texts)} 个文本区块。')

# 心理学解梦知识库向量化
texts = load_and_split_text_files(psy_knowledge_base_dir)
print(f"从心理学知识库中载入 {len(texts)} 个文本分块.")
build_vector_base(texts, vector_base_dir, psy_name)
print(f'已完成心理学向量库构建，共包含 {len(texts)} 个文本区块。')
```

### 3.2.5 检索排序

下面的代码定义了一个函数`retrieve_top_k`，它的功能是从一个向量数据库中检索与给定输入最相关的前`k`个文本。这是实现在大型文本集合中进行快速信息检索的一种有效方法，特别是在需要根据语义相似性进行文本检索的场景下。

具体来说首先成输入文本的embedding读取向量索引这里使用的是使用 FAISS库从 `vector_index` 路径读取预先构建的向量索引。这个索引包含了大量文本的向量表示，用于快速相似性搜索。

接着格式化embedding，执行相似性搜索这里使用 FAISS 的 `search` 方法，找到与输入向量在向量空间中最相似的前k个向量。

最后根据 FAISS 返回的索引，从处理好的文本集合中提取出最相关的k个文本。

通过这段代码，系统能够根据输入问题的语义内容，在庞大的知识库中快速找到最相关的信息。这是实现高效知识检索和基于内容的推荐系统的关键技术，尤其适用于需要处理大量数据和实时响应用户查询的应用场景。

```python
# 从向量库中检索topk相关文本
def retrieve_top_k(rag_input, vector_index, knowledge_base,k=10):
    rag_input_embedding = generate_embeddings(rag_input)
    index = faiss.read_index(vector_index)
    rag_input_embedding = np.array([rag_input_embedding]).astype('float32')
    distances, indices = index.search(rag_input_embedding, k)
    texts = load_and_split_text_files(knowledge_base)
    top_k_texts = [texts[i] for i in indices[0]]
    return top_k_texts

# for i, text in enumerate(top_k_texts):
#     print(f"Top {i+1} text: {text}\n")
```

### 3.2.6 重排序

下面的代码定义了一个名为 `rerank_texts` 的函数，用来实现**重排序**的功能，即用来重新排序一组文档相对于一个查询（query）的相关性。这个函数的目标是利用文本重排序模型，特别是在初步检索结果之后，希望通过更精细的评分机制来提升结果的质量。重排序在RAG中具有良好的效果已经成为了一种行业共识。

具体来说通过 `TextReRank` 类的 `call` 方法调用了阿里云的`gte_rerank`模型，这是阿里提供的重排序模型，可以根据查询的语义内容对一组文档进行评分和排序。然后解析响应体中的结果，提取每个结果的文本内容，并返回这些文本组成的列表。这个列表包含了根据查询重新排序后的最相关的文档。通过这种方法能够显著提高RAG中召回数据的质量。

```python
# 对文本进行重新排序
def rerank_texts(query, documents):
    resp = TextReRank.call(
        model=TextReRank.Models.gte_rerank,
        query=query,
        documents=documents,
        top_n=10,
        return_documents=True
    )
    if resp.status_code == HTTPStatus.OK:
        return [result['document']['text'] for result in resp.output['results']]
    else:
        print("ReRank failed:", resp)
        return documents
```

### 3.2.7 角色创建
本项目实现了4个功能：解梦功能、续写梦境功能和鼓励功能。为此需要为每个功能通过Prompt Engineering的方式根据不同的角色和情境设定了不同的模板。具体来说我设计了`meta_get_response` `psy_get_response` `sis_get_response` `writer_get_response`四个函数。

- `meta_get_response` 函数模拟一个知识渊博的传统老者的角色，提供基于提供的传统知识背景的解梦分析。

- `psy_get_response` 函数扮演心理学专家的角色，基于心理学知识库解释梦境。

- `sis_get_response` 函数作为一个温暖理解的倾听者，提供情感支持和鼓励。

- `writer_get_response` 函数作为一个作家，根据用户的梦境和提供的梦境类文档和书籍的背景续写梦境故事。

每个函数都实现了通过调用OpenAI的 API 来生成文本的功能，在这里我调用了通义千问的文本生成模型（`qwen-max`）。通过这种方法构成了一个复杂的文本生成系统，基于不同的角色和风格应对不同的用户需求，展示了使用LLM在定制化应用中的多样性和灵活性。通过精心设计的模板和策略，系统能够以不同的身份和知识背景，针对具体问题提供有针对性的答案。这种方法有效地将人工智能的能力应用于用户交互和内容生成，提高了交互的质量和相关性。

```python
def meta_get_response(question, ra_knowledge):
    template = f"""【问题】：{question}
【文档】：{', '.join(ra_knowledge)}
你是一个知识渊博的传统老者，对传统的解梦非常了解，你的任务是采用以上【文档】的内容，回答用户咨询的【问题】。请注意：
一、你的回答必须紧密依托于提供的【文档】资料，鼓励适当的语言润色来增强回答的可读性，但必须忠实于【文档】的内容。
二、回答中不要存在“根据【文档】”这类表述。
三、请你以“老夫认为”作为回答的开头。
四、你的回答风格要符合知识渊博的传统老者。
针对用户提出的【问题】：{question}，你的回答为："""
    messages = [
        # {'role': 'system', 'content': 'You are a helpful assistant.'},
        {'role': 'user', 'content': template}
    ]
    completion = openai.ChatCompletion.create(
        model="qwen-max",
        messages=messages
    )
    return completion.choices[0].message['content']


def psy_get_response(question, ra_knowledge):
    template = f"""【问题】：{question}
【文档】：{', '.join(ra_knowledge)}
你是一位心理学专家，擅长通过心理学的理论和实践来解释梦境，提供科学而深入的解析。你的任务是采用以上【文档】的内容，回答用户咨询的【问题】。请注意：
一、你的回答必须紧密依托于提供的【文档】资料，鼓励适当的语言润色来增强回答的可读性，但必须忠实于【文档】的内容。
二、回答中不要存在“根据【文档】”这类表述。
三、请你以“根据心理学理论”作为回答的开头。
四、你的回答风格要符合心理学专家的专业和深入解析。
针对用户提出的【问题】：{question}，你的回答为："""
    messages = [
        {'role': 'user', 'content': template}
    ]
    completion = openai.ChatCompletion.create(
        model="qwen-max-longcontext",
        messages=messages
    )
    return completion.choices[0].message['content']

def sis_get_response(question):
    template = f"""【问题】：{question}
你是一个充满温暖和理解的倾听者，你跟提问者是关系很好的朋友，不仅能解析梦境，还能提供非常多的安慰和鼓励。你的任务是回答用户咨询的【问题】。请注意：
一、你的回答要充满温暖和理解。
二、请提供情感支持和鼓励。
三、你的回答风格要符合倾听者的温暖和安慰风格。
针对用户提出的【问题】：{question}，你的回答为："""
    messages = [
        {'role': 'user', 'content': template}
    ]
    completion = openai.ChatCompletion.create(
        model="qwen-max",
        messages=messages
    )
    return completion.choices[0].message['content']


def writer_get_response(question, ra_knowledge):
    template = f"""【问题】：{question}
【文档】：{', '.join(ra_knowledge)}
你是一位擅长梦境续写的作家，能够结合以上【文档】的内容，续写用户的梦境。你的任务是采用以上【文档】的内容，回答用户咨询的【问题】，并续写梦境的后续。请注意：
一、你的续写内容必须紧密依托于提供的【文档】资料，但可以进行适当的创造性延展。
二、回答中不要存在“根据【文档】”这类表述。
三、请首先概述作者的提问作为续写的开头然后衔接至续写的内容。
四、你的续写风格要充满想象力和故事性。
五、请你输出的文本内容短小。仅仅需要简练的续写即可，没必要回答特别多的文字。限制在50个字以内。
针对用户提出的【问题】：{question}，你的回答为："""
    messages = [
        {'role': 'user', 'content': template}
    ]
    completion = openai.ChatCompletion.create(
        model="qwen-max",
        messages=messages
    )
    return completion.choices[0].message['content']
```

### 3.2.8 梦境助手Agent
上面的代码中为了针对具体问题提供有针对性的答案，我设计了4个角色使得能够基于提问文本内容给出更加针对性的回答。然而如何管理四个角色是一个困难的事情。我应用了**Agent**技术，通过Agent智能化的管理各个角色。我通过设计`judge_agent`函数，用于决定一个特定的梦境问题应该属于哪一个处理类别：解梦任务、鼓励任务或续写任务。

具体来说我构造一个Prompt并将其提交给LLM，从而根据梦境的内容推断出适合的响应类别。基于**few-shot learning**的思想，我的提示中包括了多个已标记的示例问题和对应的类别，通过向模型展示少量的例子来指导模型进行正确的推断。代码中我使用了通义千问的qwen-max模型。

`judge_agent`函数是本系统中的关键部分，它允许系统自动确定如何针对不同类型的梦境问题提供个性化的响应。通过这种方式，系统能够模拟不同的专家角色来处理梦境问题，从而提供更加精确和个性化的用户体验。

```python
def judge_agent(question):
    prompt = f"""
问题：我梦见自己变成了另一个人，这个梦有什么含义？
类别：解梦任务
问题：我梦见自己被朋友背叛，这让我很难过，我对他明明挺好的。
类别：鼓励任务
问题：我梦到亲人生病了，这个梦让我很害怕。
类别：鼓励任务
问题：我梦见自己在公众面前出丑，醒来后觉得很自卑，超级难过。
类别：鼓励任务
问题：我梦到自己在追求梦想的路上不断失败，我感到很失落。
类别：鼓励任务
问题：我梦见自己失去了重要的东西，这让我很焦虑，特别的伤心。
类别：鼓励任务
问题：我梦见自己在飞翔，但突然坠落，这有什么意义？
类别：解梦任务
问题：我梦到自己在黑暗的森林中迷路，这个梦代表什么？
类别：解梦任务
问题：我梦见自己和一条巨大的蛇对峙，这有什么特别的象征吗？
类别：解梦任务
问题：我梦到自己在考试中答不出来题目，这意味着什么？
类别：解梦任务
问题：我梦见自己在一个神秘的城堡中探索，故事会怎样继续？
类别：续写任务
问题：我梦到自己在宇宙中旅行，接下来会发生什么？
类别：续写任务
问题：我梦见自己找到了一本魔法书，故事会如何发展？
类别：续写任务
问题：我梦到自己变成了一只动物，接下来的冒险会怎样？
类别：续写任务
问题：我梦见自己在一场奇幻的战争中战斗，之后会发生什么？
类别：续写任务

我们的类别为：鼓励任务、解梦任务、续写任务。
现在请你对下面的这一问题分类：
问题：{question}
类别："""
    messages = [
        {'role': 'user', 'content': prompt}
    ]
    completion = openai.ChatCompletion.create(
        model="qwen-max",
        messages=messages
    )
    category = completion.choices[0].message['content'].strip()
    return category
```

### 3.2.9 角色分配

通过定义函数 `get_response_based_on_category`，实现根据用户提出的问题（梦境问题），通过上面设计的Agent判定问题所属的类别（如解梦任务、鼓励任务、续写任务），然后根据这个类别来调用相应的处理函数并返回适当的响应。

具体来说首先调用 `judge_agent` 函数来判断问题属于哪个预定义的类别。

然后使用 `if` 语句对不同的类别进行分支处理。
   - **鼓励任务**：如果分类结果为“鼓励任务”，则直接调用 `sis_get_response` 函数来生成一个鼓励性的回答。
   - **解梦任务**：
     - 使用 `retrieve_top_k` 函数从两个不同的知识库（metaphysic 和 psychology）检索与问题相关的顶部文本。
     - 将检索到的文本进行合并，然后通过 `rerank_texts` 函数重新排序以优化文本的相关性。
     - 根据重排序后文本的来源（metaphysic 或 psychology）多数，调用 `meta_get_response` 或 `psy_get_response` 生成最终答案。
   - **续写任务**：
     - 类似于解梦任务，从两个知识库检索文本，合并后重排序。
     - 调用 `writer_get_response` 函数，基于这些文本续写用户的梦境故事。

通过上面代码中复杂的决策流程的实现，为用户根据其提问**自适应**地提供针对性的、内容丰富的答案。此函数为梦境解析系统中的核心部分，实现了用户问题的自动识别和分类，并根据问题的性质提供定制化的响应。这显示了在处理复杂的用户输入时如何运用多种技术和策略来提高系统的适应性和响应质量。

```python
# 根据判断结果调用相应角色的函数
def get_response_based_on_category(question):
    while True:
        category = judge_agent(question)
        if category == "鼓励任务":
            return sis_get_response(question)
        elif category == "解梦任务":
            meta_top_k_texts = retrieve_top_k(question, 'vector_base/metaphysic_index', 'knowledge_base/metaphysic', k=10)
            psy_top_k_texts = retrieve_top_k(question, 'vector_base/psychology_index', 'knowledge_base/psychology', k=10)
            combined_texts = meta_top_k_texts + psy_top_k_texts
            reranked_texts = rerank_texts(question, combined_texts)
            # print('\n',reranked_texts,"\n")
            meta_count = sum(1 for text in reranked_texts if text in meta_top_k_texts)
            # print("meta_count:",meta_count)
            if meta_count > 5:
                return meta_get_response(question, [text for text in reranked_texts if text in meta_top_k_texts][:10])
            else:
                return psy_get_response(question, [text for text in reranked_texts if text in psy_top_k_texts][:10])
        elif category == "续写任务":
            meta_top_k_texts = retrieve_top_k(question, 'vector_base/metaphysic_index', 'knowledge_base/metaphysic', k=10)
            psy_top_k_texts = retrieve_top_k(question, 'vector_base/psychology_index', 'knowledge_base/psychology', k=10)
            combined_texts = meta_top_k_texts + psy_top_k_texts
            reranked_texts = rerank_texts(question, combined_texts)
            return writer_get_response(question, reranked_texts[:10])
        else:
            continue
```
